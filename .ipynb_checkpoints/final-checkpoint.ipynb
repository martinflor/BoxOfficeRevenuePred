# Features Engineering

```python
X1 = pd.read_csv("X1.csv")
X1.drop(columns = ["Unnamed: 0", "img_url", "studio", "description"], inplace=True)
Y1 = pd.read_csv("Y1.csv", header = None, names = ["revenue"])
Y1["revenue"] = np.log10(Y1["revenue"])

X2 = pd.read_csv("X2.csv")
X2.drop(columns = ["Unnamed: 0", "img_url", "studio", "description"], inplace=True)


# Removing Multiple occurences of a film

print("Number of films in the dataset :", len(X1["title"].values))
print("Number of unique film : ", len(np.unique(X1["title"].values)))
X1 = X1[~X1["title"].duplicated(keep="first")]

vectorizer = vectorize_title()

for i in range(len(X1)) :
    X1["title"].iloc[i] = vectorizer.transform([X1["title"].iloc[i]]).toarray()

for i in range(len(X2)) :
    X2["title"].iloc[i] = vectorizer.transform([X2["title"].iloc[i]]).toarray()



# Drop NaN values (that are stored as \\N)
for col in X1.columns : 
    X1.drop(X1[X1[col] == "\\N"].index, inplace=True)
    
for col in ['ratings', 'n_votes', 'production_year', 'runtime', 'release_year'] : 
    X2[X2[col] == "\\N"] = np.nan

# Set the type of certain features
X1["release_year"] = X1["release_year"].values.astype(np.float64)
X1["runtime"]      = X1["runtime"].values.astype(np.float64)

X2["release_year"] = X2["release_year"].values.astype(np.float64)
X2["runtime"]      = X2["runtime"].values.astype(np.float64)



cols_X2 = ['ratings', 'n_votes', 'production_year', 'runtime', 'release_year']
for col in cols_X2:
    X2[col] = X2[col].fillna(X2[col].mean())

#Reordering Indexes
Y1 = Y1.iloc[X1.index]
X1.reset_index(inplace=True)
Y1.reset_index(inplace=True)
X1.drop(columns=["index"], inplace=True)
Y1.drop(columns=["index"], inplace=True)  


# Mean embeddings : For each row in embeddings columns, the mean is taken and replace the list
#X1 = mean_embeddings(X1)


# Splitting genres features into multiple features, one for each genre.
lst_genres = list()

for i in range(len(X1["genres"])) :
    lst_genres += X1["genres"].loc[i].split(",")
    
cat_genre = np.unique(lst_genres)
X1[cat_genre] = 0 # initialize empty columns
X2[cat_genre] = 0

for i in range(len(X1)) :
    genre = X1["genres"][i]
    genre = genre.split(",")
    for j in genre :
        X1[j][i] = 1
        
for i in range(len(X2)) :
    genre = X2["genres"][i]
    if genre is np.nan:
        continue
    genre = genre.split(",")
    for j in genre :
        X2[j][i] = 1
        
X = X1[["text_embeddings", "img_embeddings"]]

X1.drop(columns = ["genres", "title", "img_embeddings", "text_embeddings", "is_adult"], inplace=True)
X2.drop(columns = ["genres", "title", "img_embeddings", "text_embeddings", "is_adult"], inplace=True)



```
